```python
import requests
import pandas as pd
```

```python
for y in range(2017, 2021):
    for m in range(1, 13):
        if m // 10 == 0:
            m = str(0)+str(m)
            
        url = 'https://nyc-tlc.s3.amazonaws.com/trip+data/yellow_tripdata_'+str(y)+'-'+str(m)+'.csv'
        r = requests.get(url, allow_redirects=True)
        name = 'tripdata_'+str(y)+'-'+str(m)+'.csv'
        open(name, 'wb').write(r.content)
        print(url)
```
